```python
first_decade= spotify_data.loc[spotify_data['year'].between(1920,1930)]
second_decade= spotify_data.loc[spotify_data['year'].between(1930,1940)]
third_decade= spotify_data.loc[spotify_data['year'].between(1940,1950)]
fourth_decade= spotify_data.loc[spotify_data['year'].between(1950,1960)]
fifth_decade= spotify_data.loc[spotify_data['year'].between(1970,1980)]
sixth_decade= spotify_data.loc[spotify_data['year'].between(1980,1990)]
seventh_decade= spotify_data.loc[spotify_data['year'].between(1990,2000)]
eigth_decade= spotify_data.loc[spotify_data['year'].between(2000,2010)]
ninth_decade= spotify_data.loc[spotify_data['year'].between(2010,2020)]
decades = [first_decade, second_decade, third_decade, fourth_decade, fifth_decade, sixth_decade, seventh_decade, eigth_decade, ninth_decade]


#data_copy = data.copy()
#indicators = pd.get_dummies(decades)
#data_copy['decade'] = data_copy
```

```python
middle_index = len(first_decade)//2
```

```python
#make this into a better formatted for loop for phase V!


for i in decades:
    sorted_by_pop = i.sort_values(by = 'popularity', ascending = False) #sort decade by decreasing popularity
    top_100 = sorted_by_pop[:middle_index] #100 most popular songs
    top_duration = top_100['duration'].mean()
    top_tempo = top_100['tempo'].mean()
    top_volume = top_100['volume'].mean()
    top_key = top_100['key'].mean()
    top_valence = top_100['valence'].mean()
    top_acoustic = top_100['acoustic'].mean()
    top_instrumental = top_100['instrumental'].mean()
    top_live = top_100['live'].mean()
    top_explicit = 1+ top_100['explicit'].mean()
    
    lower_100 = sorted_by_pop[middle_index:] #100 least popular songs
    lower_duration = lower_100['duration'].mean()
    lower_tempo =lower_100['tempo'].mean()
    lower_volume =lower_100['volume'].mean()
    lower_key =lower_100['key'].mean()
    lower_valence =lower_100['valence'].mean()
    lower_acoustic =lower_100['acoustic'].mean()
    lower_instrumental =lower_100['instrumental'].mean()
    lower_live =lower_100['live'].mean()
    lower_explicit =1+lower_100['explicit'].mean()
    
    print("Begin decade")
    print("Difference in duration: ", ((top_duration - lower_duration)/top_duration)*100)
    print("Difference in tempo: ", ((top_tempo - lower_tempo)/top_tempo)*100)
    print("Difference in volume: ", ((top_volume - lower_volume)/top_volume)*100)
    print("Difference in key: ", ((top_key - lower_key)/top_key)*100)
    print("Difference in valence: ", ((top_valence - lower_valence)/top_valence)*100)
    print("Difference in acoustic: ", ((top_acoustic - lower_acoustic)/top_acoustic)*100)
    print("Difference in instrumental: ", ((top_instrumental - lower_instrumental)/top_instrumental)*100)
    print("Difference in live: ", ((top_live - lower_live)/top_live)*100)
    print("Difference in explicit: ", ((top_explicit - lower_explicit)/top_explicit)*100)
    print("\n")
```
